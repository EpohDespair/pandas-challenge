```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
# display data set to obtain column names
purchase_data.head()
```

```python
# find number of unique players using SN column
total_players = len(purchase_data["SN"].unique())
# create dataframe to display
total_players_df = pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# find number of unique items 
unique_items = len(purchase_data["Item Name"].unique())
# calculate average price
avg_price = purchase_data["Price"].mean()
# calcluate total purchases
total_purchases = len(purchase_data["Purchase ID"].unique())
# calculate total revenue
total_revenue = purchase_data["Price"].sum()
# dataframe to display 
purchasing_df = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [avg_price], "Number of Purchases": [total_purchases], "Total Revenue": [total_revenue]})
# format
pd.options.display.float_format = '${:,.2f}'.format
purchasing_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# need to filter down to uniques for SN
filtered_purchase_data = purchase_data.drop_duplicates(subset=['SN'])
# gender breakdown and counts
gender_groupby = filtered_purchase_data.groupby(['Gender'])
# dataframe creation
gender_df = pd.DataFrame(gender_groupby["SN"].count())
gender_df.head()
#rename SN to Total Count
gender_df_renamed = gender_df.rename(columns={"SN": "Total Count"})
# add column for percent
gender_df_renamed["Percentage of Players"] = ((gender_df_renamed["Total Count"] / 576))
# format percent column
pd.options.display.float_format = '{:.2%}'.format
gender_df_renamed.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# create unfiltered gender groupby
gender_groupby2 = purchase_data.groupby(['Gender'])
# gender_groupby2.count().head()
gender_groupby2_df = pd.DataFrame(gender_groupby2['Purchase ID'].count())
# gender_groupby2_df.head()
# rename column Purchase ID
gender_groupby2_rename = gender_groupby2_df.rename(columns={"Purchase ID": "Purchase Count"})
# gender_groupby2_rename.head()
# calculate purchase total
gender_groupby2_rename["Total Purchase Value"] = gender_groupby2["Price"].sum()
#calculate Avg Total Purchase per Person
gender_groupby2_rename["Avg Total Purchase per Person"] = gender_groupby2_rename["Total Purchase Value"] / gender_df_renamed["Total Count"]
# calculate average - total purchase price/purchase count
gender_groupby2_rename["Average Purchase Price"] = (gender_groupby2_rename["Total Purchase Value"]/gender_groupby2_rename["Purchase Count"])
# gender_groupby2_rename.head()
# rearrange columns
organized_gender_df = gender_groupby2_rename[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
organized_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# create bins
bins = [0, 9, 14, 19, 24, 28, 34, 39, 100]
age_bins = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
# add Age range bins to data frame
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], bins, labels = age_bins, include_lowest = True)
# need to filter down to uniques for SN
filtered_purchase_data = purchase_data.drop_duplicates(subset=['SN'])
filtered_purchase_data.head()
# gender breakdown and counts
age_groupby = filtered_purchase_data.groupby(['Age Range'])
# data frame creation and total count
age_df = pd.DataFrame(age_groupby["SN"].count())
age_df.head()
#rename SN to Total Count
age_df_renamed = age_df.rename(columns={"SN": "Total Count"})
# add column for percent
age_df_renamed["Percentage of Players"] = ((age_df_renamed["Total Count"] / 576))
# format percent column
pd.options.display.float_format = '{:.2%}'.format
age_df_renamed
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# binning done above
# establish age groupby w/ purchase count
age_groupby_unfiltered = purchase_data.groupby(['Age Range'])
# age_groupby_unfiltered.head()
# data frame creation and total count
age_analysis_df = pd.DataFrame(age_groupby_unfiltered["SN"].count())
# age_analysis_df.head()
# rename column SN
age_analysis_df = age_analysis_df.rename(columns={"SN": "Purchase Count"})
#age_analysis_df.head()
# adjust formatting for float
pd.options.display.float_format = '${:,.2f}'.format
# calculate purchase total
age_analysis_df["Total Purchase Value"] = age_groupby_unfiltered["Price"].sum()
age_analysis_df.head()
#calculate Avg Total Purchase per Person
age_analysis_df["Avg Total Purchase per Person"] = age_analysis_df["Total Purchase Value"] / age_df_renamed["Total Count"]
# calculate average - total purchase price/purchase count
age_analysis_df["Average Purchase Price"] = (age_analysis_df["Total Purchase Value"]/age_analysis_df["Purchase Count"])
age_analysis_df.head()
# rearrange columns
org_age_analysis_df = age_analysis_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]
org_age_analysis_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# create unfiltered gender groupby
SN_groupby = purchase_data.groupby(['SN'])
#SN_groupby.count().head()
SN_groupby_df = pd.DataFrame(SN_groupby['Purchase ID'].count())
#SN_groupby_df
# rename column Purchase ID
SN_groupby_df = SN_groupby_df.rename(columns={"Purchase ID": "Purchase Count"})
#SN_groupby_df.head()
# adjust formatting for float
pd.options.display.float_format = '${:,.2f}'.format
# calculate purchase total
SN_groupby_df["Total Purchase Value"] = SN_groupby["Price"].sum()
SN_groupby_df
# calculate average - total purchase price/purchase count
SN_groupby_df["Average Purchase Price"] = (SN_groupby_df["Total Purchase Value"]/SN_groupby_df["Purchase Count"])
SN_groupby_df.head()
# rearrange columns
org_SN_group_df = SN_groupby_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
org_SN_group_df.head()
#sort Purchase Count in descending order
sort_purchase_SN = org_SN_group_df.sort_values(['Purchase Count'], ascending=False)
# display preview of data frame
sort_purchase_SN.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# create unfiltered gender groupby
item_groupby = purchase_data.groupby(['Item ID', 'Item Name'])
#item_groupby.count().head()
item_groupby_df = pd.DataFrame(item_groupby['Purchase ID'].count())
item_groupby_df.head()
# sort
item_groupby_df = item_groupby_df.sort_values(['Purchase ID'], ascending=False)
item_groupby_df.head()
# rename column Purchase ID
item_groupby_df = item_groupby_df.rename(columns={"Purchase ID": "Purchase Count"})
item_groupby_df.head()
# calculate purchase total
item_groupby_df["Total Purchase Value"] = item_groupby["Price"].sum()
item_groupby_df.head()
# add column for Item Price
item_groupby_df["Item Price"] = item_groupby_df["Total Purchase Value"] / item_groupby_df["Purchase Count"]
# reorganize
item_groupby_df = item_groupby_df[["Purchase Count", "Item Price", "Total Purchase Value"]]
item_groupby_df.head()
```
